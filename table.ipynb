```python
def reader(s):
    df = pd.read_csv(s,encoding = 'utf-16')
    df = df.values.tolist()
    all_three = 0
    inName=0
    inTitle=0
    inPara=0
    nameTitle=0
    titlePara=0
    paraName=0
    for i in df:
        if (i[10]==1.0 and i[11]==1.0 and i[12]==1.0):
            all_three = all_three + 1
        elif (i[10]==1.0 and i[11]==1.0):
            nameTitle = nameTitle + 1
        elif (i[10]==1.0 and i[12]==1.0):
            paraName = paraName + 1
        elif (i[11]==1.0 and i[12]==1.0):
            titlePara = titlePara + 1
        elif (i[10]==1.0):
            inName = inName + 1
        elif (i[11]==1.0):
            inTitle = inTitle + 1
        elif (i[12]==1.0):
            inPara = inPara + 1
        #do nothing
        
    return(all_three,inName,inTitle,inPara,nameTitle,titlePara,paraName)
    
```

```python
headers = ["Category,All_three,In Name,In Title,In Para,Name and Title,Para and Title,Para and Name"]
data = []
data.append(headers)
```

```python
s = "parsed_dataset_services.csv"
all_three,inName,inTitle,inPara,nameTitle,titlePara,paraName = reader(s)
data.append([s,all_three,inName,inTitle,inPara,nameTitle,titlePara,paraName])
data
```

```python
with open("table.csv","w") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
